```python
import pandas as pd
df = pd.read_csv('HarleyDavidsonSalesData.csv')
```

```python
df.head()
```

There is one column in particular that's missing and crucial: Profit. I will define the Profit column as Quantity x Price, and name it "Order_Profit".

```python
Order_Profit = df["Quantity"] * df["Price"]
df["Order_Profit"] = Order_Profit
```

```python
df.describe()
```

Now that I've included a Profit column, I am ready to commence the EDA. I will further analyze the columns by creating different visuals below:

```python
print(df['Order Number'].count())
print("There are 2,823 individual orders and the same number of rows in the dataset, so this looks good")
```

```python
print(df['Product Name'].unique())

count_product_name = len(df['Product Name'].unique())
print("There are", count_product_name, "series of motorcycles")
```

```python
print("The different quantity shipments come in sets of",df['Quantity'].unique())
print("There are",len(df['Quantity'].unique()), "different order amounts")
```

```python
print("The order prices per motorcyle are",df['Price'].unique())
print("There are",len(df['Price'].unique()),'different prices for the motorcycles, in $US')
```

```python
print("The different payment methods are",df['Payment Mode'].unique())
```

```python
print("There are",len(df['Store Name'].unique()), "stores in total")
```

```python
print("The shipping locations are located in these cities:",df['City'].unique())
print("The stores are located in", len(df['City'].unique()),"cities")
```

```python
print("The countries ordering motorcyle shipments are",df['Country'].unique())
print("There are", len(df['Country'].unique()), 'countries receiving motorcyle shipments')
```

```python
df['Order Date'].max()
df['Order Date'].min()
print("The first order happened on", df['Order Date'].min(), "and the last one on", df['Order Date'].max())
```

```python
print("The most profitable order was worth $", df['Order_Profit'].max(), "and the least profitable order was worth $", df['Order_Profit'].min())
```

```python
df.groupby('Country')['Order Number'].count().plot.barh()
print("The USA is responsible for the largest proportion of orders")
```

```python
print(df.groupby('Country')['Order Number'].count())
print("There were",df['Order Number'].count(), "orders in total")
print("The USA has the largest amount of orders at 1004, and Ireland the fewest with just 16")
print("The USA accounts for", round(1004*100/2823,2),"% of orders, and Ireland just", round(16*100/2823,2),"% ")
```

```python
grouped = df.groupby('Country').sum().reset_index()
grouped.sort_values('Order_Profit', ascending=False)
```

```python
grouped
```

```python
print("Harley Davidson amassed $", df.Order_Profit.sum(), "in sales")
```

```python
print("The USA accounts for", round(grouped.Order_Profit.iloc[18].sum()/(grouped.Order_Profit.sum())* 100,2), "% of Harley's profit")
print("Ireland accounts for just", round(grouped.Order_Profit.iloc[8].sum()/(grouped.Order_Profit.sum())* 100,2), "% of Harley's profit")
```

```python
df.groupby('Payment Mode').Country.count().plot.barh()
print(df.groupby('Payment Mode').Country.count())
print("Cash is the common payment method, accounting for", round(1050*100/2823, 2),"% of transactions" )
print("Loans are the least payment method, accounting for just", round(486*100/2823, 2),"% of transactions" )
```

```python
df.groupby('Product Name')['Order Number'].count().plot.barh()
print(df.groupby('Product Name')['Order Number'].count())
print("The Harley Davidson Roadster is Harley's best-selling series, accounting for", round(630*100/2823,2),"% of motorcycles sold")
print("The Harley Davidson Fat Bob is Harley's worst selling series, accounting for", round(40*100/2823,2),"% of motorcycles sold")
```

```python
df.groupby('Product Name')['Order_Profit'].sum().plot.barh()
print(df.groupby('Product Name')['Order_Profit'].sum())
print("Harley Davidson yieled $",df['Order_Profit'].sum(), "in profit during this period")
print("The Harley Davidson Street Bob is Harley's most profitable motorcycle, responsible for", round(423072000*100/1294443868,2),"% of profits")
print("The Harley Davidson Fat Boy is Harley's least profitable motorcycle, responsible for", round(12810000*100/1294443868,2),"% of profits")
```

```python
grouped.groupby('Country')['Order_Profit'].sum()
```

```python
df.groupby('Month')['Order Number'].count().plot.barh()
print(df.groupby('Month')['Order Number'].count())
print("November has the largest number of orders, with", round(329*100/2823,2),"% of the total amount")
print("September has the fewest number of orders, with only", round(168*100/2823,2),"% of the total amount")
```

```python
df.groupby('Month').Order_Profit.sum().plot.barh()
print(df.groupby('Month')['Order_Profit'].sum())
print("November yields the highest profit of any month, accounting for", round(168238277*100/1294443868,2),"% of sales")
print("September yields the lowest profit of any month, accounting for", round(74208000*100/1294443868,2),"% of sales")
```

```python

import numpy as np
import matplotlib.pyplot as plt
 
  

  
fig = plt.figure(figsize = (20, 5))
 
# creating the bar plot
plt.bar(df['City'], df['Quantity'], color ='maroon',
        width = 0.4)
 
plt.xlabel("Customer Cities")
plt.xticks(rotation=90)
plt.ylabel("No. of motorcycles sold")
plt.title("Number of motorcycle sold by city")
plt.show()
print("Strasbourg, France received more motorcycles than any other city")
```
